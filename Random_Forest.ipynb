```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn import model_selection
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv("training data.csv")
data.set_index("PDB codes",inplace=True)
```

```python
val_acc = []
train_acc = []
depth_values = np.arange(1,25,1)
for depth in depth_values:
    rf = RandomForestClassifier(criterion='gini',max_depth=depth, random_state=50)
    cv_score = model_selection.cross_validate(rf,data.iloc[:,:-1],data["Output"],cv=5,return_train_score=True)
    val_acc.append(np.mean(cv_score["test_score"]))
    train_acc.append(np.mean(cv_score["train_score"]))

plt.plot(depth_values,val_acc)
plt.plot(depth_values,train_acc)
plt.title("Random Forest with Gini Index")
plt.xlabel("max depth")
plt.ylabel("train/validation accuracy")
plt.show()
```

```python
print("optimal max depth is:",depth_values[np.argmax(val_acc)],"with validation accuracy :",val_acc[np.argmax(val_acc)])
```
