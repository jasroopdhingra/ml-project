```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn import model_selection
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv("training data.csv")
data.set_index("PDB codes",inplace=True)
train, val = model_selection.train_test_split(data, train_size=0.7)
```

```python
val.iloc[:,:-1]
```

```python
val_acc = []
train_acc = []
for k in range(1,50,2):
    knn = KNeighborsClassifier(k)
    knn.fit(train.iloc[:,:-1],train["Output"])
    pred = knn.predict(train.iloc[:,:-1])
    train_acc.append((np.sum(pred == train["Output"]))/len(pred))
    pred = knn.predict(val.iloc[:,:-1])
    val_acc.append((np.sum(pred == val["Output"]))/len(pred))

plt.plot(np.arange(1,50,2),val_acc)
plt.xlabel("K")
plt.ylabel("validation accuracy")
plt.show()
```
