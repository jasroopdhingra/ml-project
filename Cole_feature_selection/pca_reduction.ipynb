```python
import pandas as pd
import numpy as np
from sklearn import model_selection
from matplotlib import pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn import svm
from sklearn.decomposition import PCA
from sklearn import preprocessing
```

```python
data = pd.read_csv('training data.csv')
```

```python
X = data.iloc[ : , 1:-1]
y = data['Output']
```

```python
pca = PCA(n_components=3)
pca.fit(X)
X_adj = pca.transform(X)
df = pd.DataFrame(X_adj)
```

```python
fig = plt.figure()
 
# syntax for 3-D projection
ax = plt.axes(projection ='3d')
 
# plotting
ax.scatter(df[0],df[1],df[2], c=y)
ax.set_title('3 principle components')

plt.show()
```

```python
pca = PCA(n_components=2)
pca.fit(X)
X_adj = pca.transform(X)
df = pd.DataFrame(X_adj)
```

```python
plt.scatter(df[0],df[1],c=y)
```

```python
scaler = preprocessing.StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
pca = PCA(n_components=3)
pca.fit(X_scaled)
X_adj = pca.transform(X_scaled)
df = pd.DataFrame(X_adj)

fig = plt.figure()
 
# syntax for 3-D projection
ax = plt.axes(projection ='3d')
 
# plotting
ax.scatter(df[0],df[1],df[2], c=y)
ax.set_title('3 principle components')

plt.show()
```

```python
pca = PCA(n_components=2)
pca.fit(X_scaled)
X_adj = pca.transform(X_scaled)
df = pd.DataFrame(X_adj)

plt.scatter(df[0],df[1],c=y)
```

```python
svm_gaussian = svm.SVC(kernel='rbf', gamma=0.35)
scores = model_selection.cross_val_score(svm_gaussian,X_scaled,y,cv=5)
print(scores.mean())
```

```python
val_acc = []
train_acc = []
l1 = np.arange(1,27,1)
for i in l1:
    pca = PCA(n_components=i)
    pca.fit(X_scaled)
    X_adj = pca.transform(X_scaled)
    df = pd.DataFrame(X_adj)
    cv_score = model_selection.cross_validate(svm_gaussian,df,y,cv=5,return_train_score=True)
    val_acc.append(np.mean(cv_score["test_score"]))
    train_acc.append(np.mean(cv_score["train_score"]))

plt.plot(l1,val_acc,label = 'validation accuracy')
plt.plot(l1,train_acc, label = 'training accuracy')
plt.xlabel("number of features")
plt.ylabel("train/validation accuracy")
plt.legend()
plt.show()
```

```python
def generate_csv(num_pc):
    data = pd.read_csv('training data.csv')
    X = data.iloc[ : , 1:-1]
    y = data['Output']

    scaler = preprocessing.StandardScaler()
    X_scaled = scaler.fit_transform(X)

    pca = PCA(n_components=num_pc)
    pca.fit(X_scaled)
    X_adj = pca.transform(X_scaled)
    df = pd.DataFrame(X_adj)
    df["Output"] = y
    df["PDB codes"] = data["PDB codes"]
    df.set_index("PDB codes",inplace=True)

    df.to_csv("PCA_components_training.csv")

generate_csv(10)
```
